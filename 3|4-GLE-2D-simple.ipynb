## Fruit production

**target** [link](http://glx.sourceforge.net/examples/2dsimple/fruitbar.html)

![](targets/2d-simple/fruitbar.png)

**data**
```
"Weekday", "Apples", "Oranges"
"Mon",     2,        6
"Tue",     10,       16
"Wed",     3,        6
"Thu",     6,        3
"Fri",     5,        4
"Sat",     3,        7
```

````julia
Figure(font="texcmr")

title("Fruit Production")
xlabel("Weekday")
ylabel("Production [×100 pound]")

data = [2 6; 10 16; 3 6; 6 3; 5 4; 3 7]

ndays = size(data, 1)

bar(1:ndays, data[:, 1], data[:, 2], fills=["darkblue", "turquoise"])

xticks(collect(1:ndays), ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat"])

preview()
````

## Sin

**target** ([link](http://glx.sourceforge.net/examples/2dsimple/sin.html))

![](targets/2d-simple/sin.png)

````julia
Figure(font="texcmr")

x = range(-2pi, stop=2pi, length=100)
y = sin.(x)

plot(x, y, col="red", smooth=true)

gca().math = true

title!("f(x)=sin(x)", dist=0.3)

xticks([-4, -3, -2, -1, 1, 2, 3, 4]/2*pi, 
       ["-2π", "-3π/2", "-π", "-π/2", "π/2", "π", "3π/2", "2π"])
yticks([-4, -3, -2, -1, 1, 2, 3, 4]/4,
       ["-1", "-3/4", "-1/2", "-1/4", "1/4", "1/2", "3/4", "1"])

preview()
````

## Normal

**target** ([link](http://glx.sourceforge.net/examples/2dsimple/normal.html))

![](targets/2d-simple/normal.png)

````julia
Figure(font="texcmr")

title("1000 samples of N(μ=0, σ=1)")
xlabel("Value")
ylabel("Frequency")

draw = randn(1000)
hist(draw, ecol="red", fill="white", nbins=16)

xlim(-4, 4)
xticks(-4:1:4)
yticks(0:10:100)

preview()
````
