## Sin

**target** ([link](http://glx.sourceforge.net/examples/2dsimple/sin.html))

![](targets/2d-simple/sin.png)

````julia
x = range(-2pi, stop=2pi, length=100)
y = sin.(x)

f = Figure(latex=true, fontsize=8)
plot!(x, y, col="red", smooth=true)

title!(raw"$f(x)=\sin(x)$", dist=0.3)

# XXX
ax = gca()
ax.math = true
ax.xaxis.min = -2pi
ax.xaxis.max = 2pi
ax.xaxis.ticks.places = [-4, -3, -2, -1, 1, 2, 3, 4]/2*pi
ax.xaxis.tickslabels.names = [raw"$-2\pi$", raw"$-3\pi/2$", raw"$-\pi$", raw"$-\pi/2$", raw"$\pi/2$", raw"$\pi$", raw"$3\pi/2$", raw"$2\pi$"]
ax.yaxis.ticks.places = [-4, -3, -2, -1, 1, 2, 3, 4]/4
ax.yaxis.tickslabels.names  = ["-1", "-3/4", "-1/2", "-1/4", "1/4", "1/2", "3/4", "1"]

preview(f)
````
